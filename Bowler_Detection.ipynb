```python
import cv2
import numpy as np
import mediapipe as mp

mp_pose = mp.solutions.pose
pose = mp_pose.Pose()

def detect_bowler_action(landmarks):
    if landmarks:
        right_wrist = landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value]
        left_wrist = landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value]
        right_elbow = landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value]
        left_elbow = landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value]
        
        if right_wrist.y < right_elbow.y and left_wrist.y > left_elbow.y:
            return "Right-Hand Bowler - Fast Ball"
        elif left_wrist.y < left_elbow.y and right_wrist.y > right_elbow.y:
            return "Left-Hand Bowler - Fast Ball"
        elif right_wrist.y > right_elbow.y and left_wrist.y < left_elbow.y:
            return "Right-Hand Bowler - Spin Ball"
        elif left_wrist.y > left_elbow.y and right_wrist.y < right_elbow.y:
            return "Left-Hand Bowler - Spin Ball"
    return "Unknown"

def select_roi(frame):
    r = cv2.selectROI("Select Bowler", frame, fromCenter=False, showCrosshair=True)
    cv2.destroyWindow("Select Bowler")
    return r

cap = cv2.VideoCapture("C:\\Users\\chand\\Downloads\\SlowMoCric_1.mp4")
ret, frame = cap.read()
if not ret:
    print("Failed to read video.")
    cap.release()
    exit()

x, y, w, h = select_roi(frame)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    
    roi = frame[y:y+h, x:x+w]
    frame_rgb = cv2.cvtColor(roi, cv2.COLOR_BGR2RGB)
    results = pose.process(frame_rgb)
    
    action = "Unknown"
    if results.pose_landmarks:
        action = detect_bowler_action(results.pose_landmarks.landmark)
    
    cv2.putText(frame, action, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)
    cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
    
    cv2.imshow("Bowler Action Detection", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
